```python
import pandas as pd
from sqlalchemy import create_engine

# --- 1. CONFIGURAR A CONEXÃO (A mesma de antes) ---
usuario_db = 'user'
senha_db = 'password'
host_db = 'host'
porta_db = 'port'
nome_banco_db = 'database'

DATABASE_URL = f'postgresql://{usuario_db}:{senha_db}@{host_db}:{porta_db}/{nome_banco_db}'
engine = create_engine(DATABASE_URL)

print("Conexão criada.")

# --- 2. DICIONÁRIO DOS ARQUIVOS BRUTOS ---
arquivos_brutos = {
    'raw_vacinados_2022': 'data/dados_vacinados_2022.csv',
    'raw_vacinados_2023': 'data/dados_vacinados_2023.csv',
    'raw_vacinados_2024': 'data/dados_vacinados_2024.csv'
}
```

```python
print("Iniciando carga dos dados BRUTOS...")

# --- 3. EXTRAÇÃO (E) e CARGA (L) ---
for nome_tabela, caminho_arquivo in arquivos_brutos.items():
    try:
        print(f"Lendo o arquivo: {caminho_arquivo}")
        df_raw = pd.read_csv(
            caminho_arquivo, 
            encoding='utf-8',
            on_bad_lines='skip' 
        )
        
        print(f"Carregando {len(df_raw)} linhas brutos na tabela: {nome_tabela}...")
        
        # O "L" (Load) - Carregando os dados brutos, sem tratamento!
        df_raw.to_sql(
            nome_tabela,
            con=engine,
            if_exists='replace',
            index=False
        )
        print(f"Tabela {nome_tabela} carregada com sucesso.")
        
    except Exception as e:
        print(f"Erro ao processar {caminho_arquivo}: {e}")

print("\nProcesso de EL (Extração e Carga) concluído.")
```
